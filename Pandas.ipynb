```python
import numpy as np
import pandas as pd
```

```python
# create Series
s = pd.Series([1, 3, 4, np.nan, 7, 8])
s
```

```python
# create DataFrame
dates = pd.date_range('20230401', periods=6) #YYYYMMDD
dates
```

```python
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
df
```

```python
# return DataFrame index
df.index
```

```python
# return DataFrame column header
df.columns
```

```python
# sort by axis
df.sort_index(axis=1, ascending=False) # axis 1: horizontal, axis 0: vertical
```

```python
# sort by value
df.sort_values(by='C')
```

```python
# selecting columns
df['B'] # return a Series
```

```python
df.B
```

```python
df[['B', 'C']]
```

```python
# selecting rows
df[0:3] 
```

```python
# Boolian indexing
df[df['B'] > 0]
```

```python
df[df > 0]
```

```python
# setting new columns
df['E'] = [np.nan, 1, 2, 3, 4, 5]
df.loc[0:3, 'F'] = 1
df
```

```python
# drop NaN values
df.dropna(how='any') # inplace is not true by default
```

```python
# fill NaN values
df.fillna(value=999)
```

```python
# boolian mask for NaN values
pd.isna(df)
```

```python
# apply user defined function
df.apply(lambda x: x.max() - x.min())
```

```python
# concatenation
splits = [df[0:3], df[3:5], df[5:]]
splits
pd.concat(splits)
```

```python
# nerge
split1 = df.loc[:, 'A':'D']
split2 = df.loc[:, ['E', 'F']]
pd.merge(split1, split2, on=dates)
```

```python
# group by
df.fillna(value=99, inplace=True)
df.groupby(['E', 'F']).sum()
```
